# Convert GameDate to DateTime

```python
df = df.reset_index()
df.columns
```

```python
dates = df[['GameDate']]
print(dates.head())
```

```python
df['day'] = df['GameDate'].str[:2]
```

```python
df['month'] = df['GameDate'].str[2:5]
```

```python
df['year'] = df['GameDate'].str[5:]
```

```python
# replace month abbreviations with a value
mapping = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}
df.replace({'month': mapping}, inplace=True)
```

```python
df['GameDate2'] = pd.to_datetime(df[['year', 'month', 'day']])
```

```python
df.drop(['day','month','year','GameDate'], axis=1, inplace=True)
df.rename(columns={'GameDate2':'date'}, inplace=True)
df
```

```python
def regular_season_goals(name):
    df1 = df[df['player'] == name]
    df2 = df1[df1['SeasonType'] == 'REG']
    df3 = df2['g'].sum()
    df4 = df2['gp'].sum()
    return [df3, df4]
```

```python
players = ['Sean Kuraly', 'Brayden Schenn', 'Tomas Hertl']

for name in players:
    goals = regular_season_goals(name)
    print(name, goals)
```
